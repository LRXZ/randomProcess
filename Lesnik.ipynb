Сгенерируйте выборку из 1000 элементов с равномерным распределением на отрезке [0,1]. Найдите
среднее значение и выборочную дисперсию для этой выборки.

```python
import matplotlib.pyplot as plt
import numpy as np
import time

NUMBER = 1000

def generateData(number):
    np.random.seed(int(time.time()))
    return np.array(np.random.rand(number))

def meanValue(arr):
    return arr.mean()

def dispersionValue(arr):
    return np.var(arr)

def main():
    x = generateData(NUMBER)
    print(meanValue(x))
    print(dispersionValue(x))
    

if __name__ == '__main__':
    main()
```

Сгенерируйте вторую выборку из 1000 элементов с экспоненциальным распределением с параметром 1. Найдите выборочную ковариацию и коэффициент корреляции для двух полученных выборок.

```python
NUMBER_EXP = 1000

def generateDataExp(num):
    np.random.seed(int(time.time()))
    return np.array(np.random.exponential(1., num))

def mulVals(arrX, arrY):
    return arrX*arrY

def covariance(x,y):
    return np.cov(x,y)

def correlation(arrX, arrY):
    return np.corrcoef(arrX,arrY)

def main():
    x = generateDataExp(NUMBER_EXP)
    y = generateData(NUMBER_EXP)
    covXY = covariance(x,y)
    corXY = correlation(x,y) 
    print(covXY)
    print(corXY)
    print("Теперь из матриц очевидно:\n cov(x,y) = ", covXY[0][1],
          "\nТеперь из матриц очевидно:\n cor(x,y) = ", corXY[0][1])

if __name__ == '__main__' :
    main()
```

Сгенерируйте выборку из 10000 значений, распределенных нормально N(2,3). На одном графике постройте гистограмму для полученной выборки и теоретическую плотность вероятности для
распределения.

```python
NUM_NORMAL = 10000
MU = 2
SIGMA = 3
NUM_COL = 30

def generateNormal(num, mu, sigma):
    np.random.seed(int(time.time()))
    return np.random.normal(mu,sigma,num)


def normalDistribution(x, mu, sigma):
    return 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x - mu)**2 / (2 * sigma**2))


def main():
    sample = generateNormal(NUM_NORMAL, MU, SIGMA)
    x = np.arange(np.min(sample),np.max(sample), 1e-2)
    y = normalDistribution(x, MU, SIGMA)
    plt.hist(sample, NUM_COL, density=True)
    plt.plot(x,y)
    plt.show()

if __name__ == '__main__':
    main()
```

Сформируйте выборку для суммы случайных величин из предыдущей выборки. Визуализируйте
полученное распределение.

```python
NUM_DIM = 100
COEFF = 0.3



def generateExp2D(coeff, numDim):
    np.random.seed(int(time.time()))
    return np.random.exponential(coeff, (numDim, numDim))
    
def sumOfSample(x):
    return np.sum(x)

def genrateSum(x):
    ans = []
    for i in range(len(x[0])):
        ans.append(sumOfSample(x[i]))
    return ans

def main():
    x = generateExp2D(COEFF, NUM_DIM)
    vals = genrateSum(x)
    plt.plot(np.arange(100),vals)
    plt.grid = True
    plt.show()

if __name__ == '__main__':
    main()
```
